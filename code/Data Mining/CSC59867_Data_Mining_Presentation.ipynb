```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Species": [
        "Alstonia Scholaris (P2)",
        "Arjun (P1)",
        "Bael (P4)",
        "Basil (P8)",
        "Chinar (P11)",
        "Gauva (P3)",
        "Jamun (P5)",
        "Jatropha (P6)",
        "Lemon (P10)",
        "Mango (P0)",
        "Pomegranate (P9)",
        "Pongamia Pinnata (P7)"
    ],
    "Diseased": [254, 232, 118, 0, 120, 142, 345, 124, 77, 265, 272, 276],
    "Healthy": [179, 220, 0, 148, 103, 277, 279, 133, 159, 170, 287, 332]
}

df = pd.DataFrame(data)

# Stacked Bar Chart
df.plot(x="Species", kind="bar", stacked=True, figsize=(12,6), color=["red","green"])
plt.title("Diseased vs Healthy Leaves per Species (Stacked)", fontsize=14)
plt.ylabel("Number of Images")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Grouped Bar Chart
df.plot(x="Species", kind="bar", figsize=(12,6), color=["red","green"])
plt.title("Diseased vs Healthy Leaves per Species (Grouped)", fontsize=14)
plt.ylabel("Number of Images")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Pie Chart (overall healthy vs diseased)
totals = df[["Diseased","Healthy"]].sum()
plt.figure(figsize=(6,6))
plt.pie(
    totals,
    labels=totals.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=["red","green"]
)
plt.title("Overall Distribution of Diseased vs Healthy Leaves", fontsize=14)
plt.show()

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(
    df.set_index("Species"),
    annot=True, fmt="d", cmap="RdYlGn"
)
plt.title("Heatmap of Diseased vs Healthy Leaves per Species", fontsize=14)
plt.show()
```
